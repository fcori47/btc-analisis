```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Explicar que tan variable es o si sigue un patron
```

```python
df = pd.read_csv('btc.csv')
df = df.rename(columns={'Date':'date','Open':'open','High':'high','Low':'low','Close':'close','Adj Close':'adj_close','Volume':'volume'})
df.drop('adj_close',axis=1,inplace=True)
df['open'] = df.open.round(2)
df['high'] = df.high.round(2)
df['low'] = df.high.round(2)
df['close'] = df.high.round(2)
df
```

```python
df3 = pd.read_csv('prophet.csv')
df3.Date = pd.to_datetime(df3.Date)
df3.sort_values(by='Date',inplace=True)
df3.drop('Change %',axis=1,inplace=True)
df3 = df3.loc[(df3.Date < '2014-09-17')]
df3['Vol.'] = df3['Vol.'].str.replace('K','')
df3['Vol.'] = df3['Vol.'].str.replace('-','')
df3.drop(['Price'],axis=1,inplace=True)
df3 = df3.rename(columns={'Vol.':'volume'})
df3 = df3.reset_index()
df3.drop('index',axis=1,inplace=True)
df3 = df3.rename(columns={'Date':'date','Open':'open','High': 'high','Low':'low'})
df3['close'] = np.nan
df3['volume-1'] = df3.volume
df3.drop('volume',axis=1,inplace=True)
df3 = df3.rename(columns={'volume-1':'volume'})
df3['volume'] = pd.to_numeric(df['volume'].astype(str).str[:-1], errors='coerce')
df3.volume = df3.volume / 30
df3.volume = df3.volume.astype(int)
df3.replace(',','', regex=True, inplace=True)
df3.open = df3.open.astype(float)
df3.high = df3.high.astype(float)
df3.low = df3.low.astype(float)
df3['close'] = df3['low']
df3
```

```python
frames = [df3,df]
bc = pd.concat(frames)
bc['date'] = pd.to_datetime(bc.date)
bc
```
